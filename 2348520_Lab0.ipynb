## (1) Generate and plot the following signals.

### A unit step function.

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.linspace(0,10,100)
y=np.heaviside(time,1)
plt.plot(time,y)
plt.stem([0],[1],'r')
```

### A unit impulse function.

```python
time=np.linspace(-5,5,100)
y=np.where(time==0,1,0)
plt.plot(time,y)
plt.stem([0],[1],'r')
plt.xlim(-5,5)
plt.ylim(-.2,1.2)
plt.show()
```

### A unit ramp function.

```python
time=np.linspace(0,10,100)
y=np.where(time>=0,time,0)
plt.plot(time,y)
```

### An exponential signal (Growing)

```python
time=np.linspace(0,10,100)
y=np.exp(time)
plt.plot(time,y)
```

### An exponential signal (Decaying)

```python
time=np.linspace(0,10,100)
y=np.exp(-time)
plt.plot(time,y)
```

### A sinusoidal signal.

```python
time=np.linspace(0,10,100)
y=np.sin(time)
plt.plot(time,y)
```

## (2) You are asked to visualize the effects of sampling and reconstructing a continuous-time signal.

### (a) Generate a continuous sinusoidal signal.

```python
time=np.linspace(0,1,100)
frequency=3
continuous_signal=np.sin(2*np.pi*frequency*time)
plt.plot(time,continuous_signal)
```

### (b) Sample the signal at different rates (Nyquist rate, above, and below Nyquist).

```python
nyquist=2*frequency
nyquistb=0.5*frequency
nyquista=4*frequency
continuous_nyquist=np.sin(2*np.pi*nyquist*time)
continuous_nyquistb=np.sin(2*np.pi*nyquistb*time)
continuous_nyquista=np.sin(2*np.pi*nyquista*time)
plt.plot(time,continuous_nyquist)
plt.plot(time,continuous_nyquistb)
plt.plot(time,continuous_nyquista)
plt.legend(['nyquist','nyquistb','nyquista'],bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

### (c) Reconstruct the sampled signal and observe the aliasing effect when un-dersampled.

```python
plt.plot(time,continuous_signal, label="Continuous Signal",alpha=0.5)
plt.stem(time,continuous_nyquistb,linefmt='r-',markerfmt='ro',basefmt=" ",label="Sampled Below Nyquist Rate")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
