## Lab Exercise 2 - Fourier Transform and Frequency Spectrum Analysis of Signals

### Q1 (a) Generate a basic sinusoidal signal in the time domain.

```python
import numpy as np
fs=1000
time=np.arange(0,1,1/fs)
frequency=5
signal=np.sin(2*np.pi*frequency*time)
```

### (b) Plot the time-domain waveform of the signal.

```python
import matplotlib.pyplot as plt
plt.plot(time,signal)
```

### (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the conftinuous frequency spectrum.

```python
from scipy.fftpack import fft,fftshift
dtft=fftshift(fft(signal,1024))
freq=np.linspace(-fs/2,fs/2,len(dtft))
plt.plot(freq,np.abs(dtft))
plt.xlim(0,10)
```

### (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
dft=fft(signal)
freq_dft=np.fft.fftfreq(len(signal),1/fs)
plt.stem(freq_dft,np.abs(dft))
plt.xlim(0,10)
```

### Q2 (a) Generate a composite signal by adding two or more sinusoidal signals of different frequencies and amplitudes.

```python
f1=5
f2=25
signal1=np.sin(2*np.pi*f1*time)
signal2=0.5*np.sin(2*np.pi*f2*time)
composite=signal1+signal2
```

### (b) Plot the time-domain waveform of the composite signal.

```python
plt.plot(time,composite)
```

### (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
dtft=fftshift(fft(composite,1024))
freq=np.linspace(-fs/2,fs/2,len(dtft))
plt.plot(freq,np.abs(dtft))
plt.xlim(0,30)
```

### (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
dft=fft(composite)
freq_dft=np.fft.fftfreq(len(composite),1/fs)
plt.stem(freq_dft,np.abs(dft))
plt.xlim(0,30)
```

### Q4 (a) Generate a Rectangle signal.

```python
fs=1000
t=np.linspace(0,10,1000) 
start=2
end=4
rect_signal1=np.where((t>=start) & (t<=end),1,0)
start=5
end=8
rect_signal2=np.where((t>=start) & (t<=end),1,0)
rect_signal=rect_signal1+rect_signal2
```

### (b) Plot the time-domain waveform.

```python
plt.plot(t,rect_signal)
```

### (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
dtft=fftshift(fft(rect_signal,1024))
freq=np.linspace(-fs/2,fs/2,len(dtft))
plt.plot(freq,np.abs(dtft))
plt.xlim(0,10)
```

### (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

### (e) Analyze the relationship between the time-domain waveform and the frequency-domain representation.

#### The rectangular signal has sharp transitions in the time domain, which introduce high-frequency components in the frequency domain. 

### Write an inference on the frequency domain analysis of various signals using DTFT and DFT.

#### A sinusoidal signal is simple in the frequency domain, with all its energy concentrated at one frequency. Both DTFT and DFT efficiently capture this characteristic.
#### A composite signal is represented by multiple frequency components, each corresponding to the individual sinusoids. Both DTFT and DFT effectively capture the multi-frequency nature of the signal.
#### The decaying exponential signal is a smooth, non-periodic signal in the time domain, and its frequency content is broad but concentrated at low frequencies.
#### The rectangular signal has sharp transitions in the time domain, which introduce high-frequency components in the frequency domain. 
#### In all cases, the time-domain waveform determines the distribution of energy in the frequency domain. Smooth, slowly varying signals like sinusoids and exponentials concentrate energy at low frequencies, while signals with sharp transitions (such as rectangular pulses) introduce higher frequency components. Both DTFT and DFT allow for effective analysis of signals in the frequency domain, but the DTFT offers a continuous representation, whereas the DFT provides a discrete, sampled version.
